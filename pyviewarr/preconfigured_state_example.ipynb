# Preconfigured pyviewarr state

Initialize a `pyviewarr.Widget` with predefined viewer state from Python.

```python
import numpy as np
import pyviewarr
from pyviewarr import ViewerConfig
```

```python
ny, nx = 256, 320
y, x = np.mgrid[-1:1:complex(0, ny), -1.5:1.5:complex(0, nx)]
data = np.sin(8 * x) * np.cos(6 * y) + 0.3 * x

config = ViewerConfig(
    contrast=1.8,
    bias=0.45,
    stretch="log",
    cmap="Inferno",
    colormap_reversed=False,
    vmin=-1.2,
    vmax=1.6,
    xlim=(40, 260),
    ylim=(30, 210),
    zoom=1.3,
    rotation=18.0,
    pivot=(160.0, 128.0),
    show_pivot_marker=True,
)
```

```python
config
```

```python
w = pyviewarr.viewarr(data, viewer_config=config, width=900, height=540)
w
```
